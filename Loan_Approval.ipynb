```python
df.head()
```

```python
df['Loan_Status']=df['Loan_Status'].map({'Y':1 , 'N': 0})
df['Self_Employed']=df['Self_Employed'].map({'Yes':1, 'No':0})
df['Married']=df['Married'].map({'Yes':1,'No':0})
df['Education']=df['Education'].map({'Graduate':1,'Not Graduate':0})
df['Gender']=df['Gender'].map({'Male':1,'Female':0})
df['Property_Area']=df['Property_Area'].map({'Urban':2,'Rural':0,'Semiurban':1})

df['Dependents']=df['Dependents'].map({'1':1,'3+': 4, '2':2,'0':0})
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['Loan_ID', 'Loan_Status'])
Y = df['Loan_Status']

# Scale numeric features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])
```

```python
print(Y)
```

```python
X_train,X_test, Y_train,Y_test=train_test_split(X,Y, test_size=0.1,stratify=Y, random_state=5)
```

```python
print(Y)
```

```python
print(X)
```

```python
print(df['Dependents'].unique())
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
model=svm.SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

```python
training= model.predict(X_train)
train_acc=accuracy_score(training,Y_train)
print("The training data accuracy is : ",train_acc)

testing=model.predict(X_test)
testing_acc=accuracy_score(testing,Y_test)
print("The testing data accuracy is : ",testing_acc)
```

```python
import pickle

# Suppose your model is called 'model'
with open('loan_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
